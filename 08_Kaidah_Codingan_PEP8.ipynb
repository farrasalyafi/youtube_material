```python
#contoh penggunaan blank lines dalam function
def calculate_variance(number_list):
    sum_list = 0
    for number in number_list:
        sum_list = sum_list + number
    mean = sum_list / len(number_list)
                                                   #ini memudahkan dalam mengetahui cara kerja functionya
    sum_squares = 0
    for number in number_list:
        sum_squares = sum_squares + number**2
    mean_squares = sum_squares / len(number_list)

    return mean_squares - mean**2
```

# 6. Whitespace

Whitesapce itu ya spasi atau ruang kosong di antara codingan kita agar codinganya gampang dibaca.Rekomendasi menggunakan white space pada operasi berikut:
1. Assignment operators (=, +=, -=, and so forth)
2. Comparisons (==, !=, >, <. >=, <=) and (is, is not, in, not in)
3. Booleans (and, not, or)

```python
#Contoh penggunaan whitepace
#Direkomendasikan
y = x**2 + 5
z = (x+y) * (x-y)

#Tidak direkmondasikan
y = x ** 2 + 5
z = (x + y) * (x - y)
```

```python
#Contoh penggunaan whitespace
#Direkomendasikan
def fungsi(parameter=5):
    #.....

#Tidak direkomendasikan
def fungsi(parameter = 5):
    #.....
```

```python
#Contoh whitespace pada if statement
#Direkomendasikan
if x>5 and x%2==0:
    print('x lebih besar dari 5 dan x dibagi 2!')

#Tidak direkomendasikan
if x > 5 and x % 2 == 0:
    print('x lebih besar dari 5 dan x dibagi 2!')
```
